````python
# Importando os pacotes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importando a base
df = pd.read_csv("previsao_de_renda.csv")
````

````python
# Eliminando as variáveis irrelevantes
df.drop(['Unnamed: 0',
         'index'], 
        axis=1, inplace=True)

# Transformando data em formato de data
df.loc[:,'data_ref'] = pd.to_datetime(df['data_ref'])

# Renomeando Educação (para organizar melhor o gráfico)
educ = {'Primário': '1. Primário',
        'Secundário': '2. Secundário',
        'Superior incompleto': '3. Superior Incompleto',
        'Superior completo': '4. Superior Completo',
        'Pós graduação': '5. Pós Graduação'}
df['educacao'] = df['educacao'].map(educ)
````

### 1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
ax = pd.crosstab(df['data_ref'],df['sexo'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), loc=2, borderaxespad=0.)
ax.set_title('Gênero')
````

````python
ax = pd.crosstab(df['data_ref'],df['posse_de_veiculo'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Posse de Veículo')
````

````python
ax = pd.crosstab(df['data_ref'],df['posse_de_imovel'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Posse de Imóvel')
````

````python
ax = pd.crosstab(df['data_ref'],df['tipo_renda'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Tipo de Renda')
````

````python
ax = pd.crosstab(df['data_ref'],df['educacao'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Educação')
````

````python
ax = pd.crosstab(df['data_ref'],df['estado_civil'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Estavo Cívil')
````

````python
ax = pd.crosstab(df['data_ref'],df['tipo_residencia'], normalize='index').plot.bar(stacked=True)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs)
plt.legend(bbox_to_anchor=(1.025, 1), reverse=True)
ax.set_title('Tipo de Residência')
````

#### <span style="color:blue;"> Conclusão:</span>

A única variável que apresentou uma mudança notável ao longo do tempo foi **educação**, em que a partir de **setembro de 2015** houve um aumento expressivo de clientes com **ensino superior completo** em percentual na base.

### 2. Para cada variável quantitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?
